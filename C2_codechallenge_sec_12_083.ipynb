```python
import numpy                as np
import sympy                as sym
import matplotlib.pyplot    as plt
import matplotlib.colors    as mcolors
import scipy.integrate      as spi
import math
import mpmath
import plotly.graph_objects as go
import sympy.stats

from scipy.signal                     import find_peaks
from scipy                            import stats
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Pdfs using numpy

# Params
x     = np.linspace(-5,5,1001)
dx    = x[1]-x[0]
sigma = 1.4

# Pdf
pdf = 1/np.sqrt(2*np.pi*sigma**2) * np.exp( -x**2/(2*sigma**2) )
pdf *= dx

# Confirm normalisation condition
sumPdf = np.sum(pdf)
print('sumPdf = %.2f' %sumPdf)

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

plt.plot(x,pdf)
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

plt.gca().set(xlabel='x',ylabel='Probability',title='Gaussian (sum = %.2f)' %sumPdf)
plt.tight_layout()

plt.savefig('fig1_codechallenge_83_exercise_0a.png')
plt.show()
files.download('fig1_codechallenge_83_exercise_0a.png')
```

```python
# %% Exercise 0a
#    Pdfs using sympy

# Params
x     = sym.symbols('x')
dfs   = [5,30]
fvals = np.linspace(.01,5,500)
dx    = fvals[1]-fvals[0]

# Sympy F distribution
F         = sym.stats.FDistribution('F',dfs[0],dfs[1])
Fpdf_expr = sym.stats.density(F)(x)

# Lambdify
Fpdf_l = sym.lambdify(x,Fpdf_expr)

# Numerically evaluate
pdf    = Fpdf_l(fvals) * dx
sumPdf = np.sum(pdf)

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

plt.plot(fvals,pdf)
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

plt.gca().set(xlim=fvals[[0,-1]],xlabel='F',ylabel='Probability',title=r'F-distribution ($d_1,d_2$ = %g,%s, sum = %.2f)' %(dfs[0],dfs[1],sumPdf))
plt.tight_layout()

plt.savefig('fig2_codechallenge_83_exercise_0b.png')
plt.show()
files.download('fig2_codechallenge_83_exercise_0b.png')
```

```python
# %% Exercise 0a
#    Pdfs using scipy

# Params
mu    = 1
sigma = .3

# Pdf values
xx = np.linspace(-5,5,401)
pdf = stats.logistic.pdf(xx,loc=mu,scale=sigma) * (xx[1]-xx[0])

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

plt.plot(xx,pdf)
plt.axvline(mu,linewidth=1,color='gray',linestyle='--')
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

plt.gca().set(xlim=xx[[0,-1]],xlabel='x',ylabel='probability',title=r'Logistic pdf ($\mu = %g, \sigma = %g$)' %(mu,sigma))
plt.tight_layout()

plt.savefig('fig3_codechallenge_83_exercise_0c.png')
plt.show()
files.download('fig3_codechallenge_83_exercise_0c.png')
```
