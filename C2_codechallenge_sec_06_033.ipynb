```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Indefinite integration in sympy

# Function
x = sym.symbols('x')
f = 4*x

# Antiderivative
F = sym.integrate(f,x)
F_display = sym.Integral(f)

display(Math( '%s = %s + C' % (sym.latex(F_display),sym.latex(F)) ))
```

```python
# %% Exercise 0b
#    Definite integration in sympy

# Function
f = x**2 + 4*x

# Indefinite integral and integration bounds
F = sym.integrate(f,x)
F_display = sym.Integral(f)

display(Math( '%s = %s + C' % (sym.latex(F_display),sym.latex(F)) ))

a = -2
b = 1

# Definite integral
F_def = sym.integrate(f,(x,a,b))
F_def_display = sym.Integral(f,(x,a,b))

display(Math( '%s = %s' % (sym.latex(F_def_display),sym.latex(F_def)) ))
```

```python
# %% Exercise 0c
#    Definite integration in sympy vis FTC

# Function
f = x**2 + 4*x

# Indefinite integral and integration bounds
F = sym.integrate(f,x)
F_display = sym.Integral(f)

display(Math( '%s = %s + C' % (sym.latex(F_display),sym.latex(F)) ))

a = -2
b = 1

# Definite integral via FTC
A = F.subs(x,b) - F.subs(x,a)

display(Math( 'A = %s - %s = %s' % (sym.latex(F.subs(x,b)),sym.latex(F.subs(x,a)),sym.latex(A)) ))
```

```python
# %% Exercise 0d
#    Visualise function and antiderivative

# Function and integral
f = 4
F = sym.integrate(f,x)

# Lambdify
f_lamb = sym.lambdify(x,f,'numpy')
F_lamb = sym.lambdify(x,F,'numpy')
xx = np.linspace(a-1,b+1,301)
xi = np.linspace(a,b,201)

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(5*phi,5))

plt.fill_between(xi,f_lamb(xi),alpha=.3,label='Area of interest')

plt.plot(xx,f_lamb(xx)*np.ones(len(xx)),label=r'f(x) = $%s$' %sym.latex(f))
plt.plot(xx,F_lamb(xx),label=r'$\int f(x) dx=%s$' %sym.latex(F))

plt.axvline(a,linestyle=':',color='r')
plt.axvline(b,linestyle=':',color='r')
plt.axhline(0,linestyle=':',color='grey')

plt.plot(a,F.subs(x,a),'ro',label=r'$F(%g) = %.2f$' %(a,F.subs(x,a)))
plt.plot(b,F.subs(x,b),'bo',label=r'$F(%g) = %.2f$' %(b,F.subs(x,b)))

plt.title('A function and its antiderivative')
plt.xlabel('x')
plt.ylabel(r'f(x) or $\int f(x)dx$')
plt.xlim(xx[[0,-1]])
plt.legend()

plt.savefig('fig1_codechallenge_33_exercise_0d.png')
plt.show()
files.download('fig1_codechallenge_33_exercise_0d.png')
```
