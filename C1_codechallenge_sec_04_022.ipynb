```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# %% Exercise 4
#    Show convergence of various Riemann sums' rules as dx decreases

# Function
def f(x):
    return np.exp(x)/10 + np.cos(x)

# Bounds and ns
a = -0.5
b = np.pi
N = np.arange(5,191,5)

areas = np.zeros((len(N),3))

# Analytic solution
x     = sym.symbols('x')
f_sym = sym.exp(x)/10 + sym.cos(x)
analytic_solution = sym.integrate(f_sym,(x,a,b))

# Dx values
for i,n in enumerate(N):

    delta_x = (b-a)/n
    delta_border = np.array([a+delta_x*i for i in range(n+1)])

    areas[i,0] = np.sum(f(delta_border[:-1])) * delta_x
    areas[i,1] = np.sum(f(delta_border[1:])) * delta_x
    areas[i,2] = np.sum(f((delta_border[:-1]+delta_border[1:])/2)) * delta_x

# Plotting
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(5*phi,5))

plt.plot(N,areas,linewidth=1.5)
plt.axhline(y=analytic_solution,color='k',linestyle=':',linewidth=1.2)
plt.legend(['Left','Right','Midpoint','Analytic'])
plt.xlim(N[[0,-1]])
plt.xlabel('Number of bins')
plt.ylabel('Area under curve')
plt.title('Convergence of Riemann sums')

plt.savefig('fig3_codechallenge_22_exercise_4.png')
plt.show()
files.download('fig3_codechallenge_22_exercise_4.png')
```
