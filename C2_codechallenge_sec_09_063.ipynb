```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Show that integration by parts works by comparing with sympy

# Function
x = sym.symbols('x')
C = sym.symbols('C')

f = x*sym.cos(x)
F = sym.integrate(f)

# Display
display(Math('%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F))))

# Discretise
xx = np.linspace(-2*np.pi,3*np.pi,501)
yy = [f.subs(x,xi) for xi in xx]
Fx = [F.subs(x,xi) for xi in xx]

# Plotting
phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,2,figsize=(1.5*phi*5,5))

axs[0].plot(xx,yy,linewidth=2,label='f')
axs[0].plot(xx,Fx,linewidth=2,label='F')
axs[0].set(xlim=xx[[0,-1]],xlabel='x',ylabel='y=f(x) or F(x)')
axs[0].set_title('Function and its antiderivative')
axs[0].legend()

axs[1].plot(yy,Fx,'k',linewidth=2)
axs[1].set(xlabel='f(x)',ylabel='F(x)')
axs[1].set_title('Function vs. antiderivative')

plt.tight_layout()

plt.savefig('fig3_codechallenge_63_exercise_0a.png')
plt.show()
files.download('fig3_codechallenge_63_exercise_0a.png')
```

```python
# %% Exercise 0b
#    Show that integration by parts works by comparing with sympy

# Function
x = sym.symbols('x')
C = sym.symbols('C')

f = sym.exp(-x*sym.pi)*sym.sin(7*x)
F = sym.integrate(f)

# Display
display(Math('%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F))))

# Discretise
xx = np.linspace(-2*np.pi,1*np.pi,501)
yy = [f.subs(x,xi) for xi in xx]
Fx = [F.subs(x,xi) for xi in xx]

# Plotting
phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,2,figsize=(1.5*phi*5,5))

axs[0].plot(xx,yy,linewidth=2,label='f')
axs[0].plot(xx,Fx,linewidth=2,label='F')
axs[0].set(xlim=xx[[0,-1]],xlabel='x',ylabel='y=f(x) or F(x)')
axs[0].set_title('Function and its antiderivative')
axs[0].legend()

axs[1].plot(yy,Fx,'k',linewidth=2)
axs[1].set(xlabel='f(x)',ylabel='F(x)')
axs[1].set_title('Function vs. antiderivative')

plt.tight_layout()

plt.savefig('fig4_codechallenge_63_exercise_0b.png')
plt.show()
files.download('fig4_codechallenge_63_exercise_0b.png')
```
