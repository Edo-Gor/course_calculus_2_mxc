```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Show that U-substitution works by comparing with sympy

# Function
x = sym.symbols('x')
C = sym.symbols('C')
f = x / (x**2 + 2)

# Integral with sympy (you can see the latex code by printing only the part :
# '%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F)) )
F = sym.integrate(f)
display(Math('%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F))))

# Check our answer match with sympy's
my_ans = 1/2 * sym.log(x**2 + 2)
display(sym.expand(my_ans))

# Plotting
xx = np.linspace(-2,5,501)
yy = [f.subs(x,xi) for xi in xx]
Fx = [F.subs(x,xi) for xi in xx]

phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,figsize=(phi*5,5))

axs.plot(xx,yy,linewidth=2,label=r'f(x) = $%s$'%sym.latex(f))
axs.plot(xx,Fx,linewidth=2,label=r'F(x) = $%s+0$'%sym.latex(F))
axs.axhline(0,linestyle=':',color='gray',zorder=-3)
axs.axvline(0,linestyle=':',color='gray',zorder=-3)

axs.set(xlim=xx[[0,-1]],xlabel='x',ylabel='f(x) or F(x)')
axs.legend()
plt.title("A function and its antiderivative")

plt.savefig('fig1_codechallenge_62_exercise_0a.png')
plt.show()
files.download('fig1_codechallenge_62_exercise_0a.png')
```

```python
# %% Exercise 0b
#    Show that U-substitution works by comparing with sympy

# Function
x = sym.symbols('x')
C = sym.symbols('C')
f = (3*x**4+5)**6 * x**3

# Integral with sympy (you can see the latex code by printing only the part :
# '%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F)) )
F = sym.integrate(f)
display(Math('%s = %s+C' %(sym.latex(sym.Integral(f)),sym.latex(F))))

# Check our answer match with sympy's
my_ans = (3*x**4+5)**7/84
display(sym.expand(my_ans))

# Plotting
xx = np.linspace(-2,5,501)
yy = [f.subs(x,xi) for xi in xx]
Fx = [F.subs(x,xi) for xi in xx]

phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,figsize=(phi*5,5))

axs.plot(xx,yy,linewidth=2,label=r'f(x) = $%s$'%sym.latex(f))
axs.plot(xx,Fx,linewidth=2,label=r'F(x) = $%s+0$'%sym.latex(F))
axs.axhline(0,linestyle=':',color='gray',zorder=-3)
axs.axvline(0,linestyle=':',color='gray',zorder=-3)

axs.set(xlim=xx[[0,-1]],xlabel='x',ylabel='f(x) or F(x)')
axs.legend()
plt.title("A function and its antiderivative")

plt.savefig('fig2_codechallenge_62_exercise_0b.png')
plt.show()
files.download('fig2_codechallenge_62_exercise_0b.png')
```
