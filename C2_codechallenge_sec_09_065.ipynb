```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 1.0
#    Visualise the function

# Function
x = sym.symbols('x')
A = sym.symbols('A')
B = sym.symbols('B')
f = (5*x+3) / (2*x**2 - 4*x - 6)
F = sym.integrate(fx)

# Plot with sympy
p = sym.plot(f,(x,-5,5),F,ylim=[-20,20],title='A function and its derivative',ylabel='f(x) / F(x)',xlabel='x',legend=True,label=['Function', 'Integral'],show=False)
p._backend.process_series()
lines = p._backend.ax.get_lines()
lines[0].set_linewidth(1.5)
lines[1].set_linewidth(1.5)
lines[1].set_linestyle('--')

p._backend.fig.savefig('fig5_codechallenge_65_exercise_1.png')
files.download('fig5_codechallenge_65_exercise_1.png')

display(Math('f(x) = %s' %sym.latex(f)))
print('')
display(Math('\\int f(x) \\,dx = ?'))
```

```python
# %% Exercise 1.1
#    Separate numerator and denominator

numerator, denominator = f.as_numer_denom()

display(Math('\\text{The numerator is } %s' %sym.latex(numerator)))
display(Math('\\text{The denominator is } %s' %sym.latex(denominator)))
```

```python
# %% Exercise 1.2
#    Factor the denominator

den_factors = sym.factor(denominator)

for i,fact in enumerate(den_factors.args):
  display(Math('\\text{Simple factors } %g: \\; %s' %(i+1,sym.latex(fact))))
  print()
```

```python
# %% Exercise 1.3
#    Create the simple fractions

simple_fract_1 = A / (den_factors.args[0]*den_factors.args[1])
simple_fract_2 = B / den_factors.args[2]

display(Math('\\text{Factor 1:} \\; %s' %sym.latex(simple_fract_1))), print()
display(Math('\\text{Factor 2:} \\; %s' %sym.latex(simple_fract_2)))
```

```python
# %% Exercise 1.4
#    Solve for A and B

expression  = sym.Eq(numerator, simple_fract_1*sym.prod(den_factors.args) + simple_fract_2*sym.prod(den_factors.args))
solutionsAB = sym.solve(expression,(A,B))

display(Math('\\text{The solutions are:}')), print()
display(Math('\\text{A = } %s' %sym.latex(solutionsAB[A]))), print()
display(Math('\\text{B = } %s' %sym.latex(solutionsAB[B])))
```

```python
# %% Exercise 1.5
#    Integrate separately

defint1 = sym.integrate( simple_fract_1.subs(A,solutionsAB[A]) )
defint2 = sym.integrate( simple_fract_2.subs(B,solutionsAB[B]) )

display(Math('\\int %s \\,dx = %s+C' %(sym.latex(simple_fract_1.subs(A,solutionsAB[A])),sym.latex(defint1)))),print()
display(Math('\\int %s \\,dx = %s+C' %(sym.latex(simple_fract_2.subs(B,solutionsAB[B])),sym.latex(defint2))))
```

```python
# %% Exercise 1.6
#    Sum the parts

my_solution = defint1 + defint2
display(Math('\\int %s \\, dx = %s+C' %(sym.latex(f),sym.latex(my_solution))))
```
