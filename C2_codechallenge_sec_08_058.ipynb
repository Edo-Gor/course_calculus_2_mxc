```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 1
#    Use scipy to compute the definite integral shown in the video, plot the
#    function and the extimation of the definite integral over various values

# Function
x = sym.symbols('x')
f = sym.cos(x)

# Lambdify (twice for net and total areas)
fx_l_net = sym.lambdify(x,f)
fx_l_tot = sym.lambdify(x,sym.Abs(f))

# Define the upper integration limits
bounds = np.linspace(np.pi/4,8*np.pi,501)

# Compute deifnite integrals over bounds dx
def_int_net = np.zeros(len(bounds))
def_int_tot = np.zeros(len(bounds))

for i,U in enumerate(bounds):
    def_int_net[i],_ = spi.quad(fx_l_net,0,U)
    def_int_tot[i],_ = spi.quad(fx_l_tot,0,U)

# Plotting
phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,2,figsize=(1.5*phi*5,5))

xx = np.linspace(0,bounds[-1],1001)
axs[0].plot(xx,fx_l_net(xx),linewidth=1.2)
axs[0].axhline(0,linestyle=':',color='grey',zorder=-3,linewidth=0.8)
axs[0].set(xlabel='Angle (x)',ylabel=r'$y = %s$'%sym.latex(f),xlim=[0,bounds[-1]],title='Function')

axs[1].plot(bounds,def_int_net,label='Net area',linewidth=1.2)
axs[1].plot(bounds,def_int_tot,label='Total area',linewidth=1.2)
axs[1].axhline(0,linestyle=':',color='grey',zorder=-3,linewidth=0.8)
axs[1].set(xlabel='Upper bound',ylabel='Area',xlim=bounds[[0,-1]],title='Areas')
axs[1].legend()

plt.tight_layout()

plt.savefig('fig3_codechallenge_58_exercise_1.png')
plt.show()
files.download('fig3_codechallenge_58_exercise_1.png')
```

```python
# %% Exercise 1
#    Repeat exercise 1 but with another function

# Function
x = sym.symbols('x')
f = sym.cos(x**2)

# Lambdify (twice for net and total areas)
fx_l_net = sym.lambdify(x,f)
fx_l_tot = sym.lambdify(x,sym.Abs(f))

# Define the upper integration limits
bounds = np.linspace(np.pi/4,8*np.pi,1001)

# Compute deifnite integrals over bounds dx (trapezoidal rule)
def_int_net = np.zeros(len(bounds))
def_int_tot = np.zeros(len(bounds))

for i,U in enumerate(bounds):
    xx_trap = np.linspace(0, U, 1001)
    def_int_net[i] = spi.trapezoid(fx_l_net(xx_trap), xx_trap)
    def_int_tot[i] = spi.trapezoid(fx_l_tot(xx_trap), xx_trap)

# Plotting
phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,2,figsize=(1.5*phi*5,5))

xx = np.linspace(0,bounds[-1],1001)
axs[0].plot(xx,fx_l_net(xx),linewidth=1.2)
axs[0].axhline(0,linestyle=':',color='grey',zorder=-3,linewidth=0.8)
axs[0].set(xlabel='Angle (x)',ylabel=r'$y = %s$'%sym.latex(f),xlim=[0,bounds[-1]],title='Function')

axs[1].plot(bounds,def_int_net,label='Net area',linewidth=1.2)
axs[1].plot(bounds,def_int_tot,label='Total area',linewidth=1.2)
axs[1].axhline(0,linestyle=':',color='grey',zorder=-3,linewidth=0.8)
axs[1].set(xlabel='Upper bound',ylabel='Area',xlim=bounds[[0,-1]],title='Areas (with trapezoid rule)')
axs[1].legend()

plt.tight_layout()

plt.savefig('fig4_codechallenge_58_exercise_2.png')
plt.show()
files.download('fig4_codechallenge_58_exercise_2.png')
```
