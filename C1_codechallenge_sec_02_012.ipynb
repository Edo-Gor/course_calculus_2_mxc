```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# %% Exercise 0a
#    Show that the integral of a derivative of a function is that function

x,C = sym.symbols('x,C')
f   = (x-1)**2
df  = sym.diff(f,x)
id = sym.integrate(df,x) + C

display(Math("\\text{Function: } %s" %sym.latex(f)))
display(Math("\\text{Derivative: } %s" %sym.latex(df)))
display(Math("\\text{Integral: } %s" %sym.latex(id)))
```

```python
# %% Exercise 0b
#    Show that F(f(x)) = f(x), when accounting for the constant

display(Math("\\text{Function: } %s" %sym.latex(sym.expand(f))))
display(Math("\\text{Integral of derivative: } %s" %sym.latex(id)))
```

```python
# %% Exercise 0c
#    Quick dirty plotting of the function

p = sym.plot(f,(x,-1,2),id.subs(C,1),title='A function and its inverse integral',ylabel='f(x) / F(x)',xlabel='x',show=False)
p._backend.process_series()
lines = p._backend.ax.get_lines()
lines[0].set_linewidth(5)
lines[1].set_linestyle('--')

p._backend.fig.savefig('fig11_codechallenge_12_exercise_0c.png')
files.download('fig11_codechallenge_12_exercise_0c.png')
```

```python
# %% Exercise 0d
#    More refined plotting

# Lambdify
f_l    = sym.lambdify(x,f)
df_l   = sym.lambdify(x,df)
id_l   = sym.lambdify(x,id.subs(C,1))
id_l_c = sym.lambdify(x,id.subs(C,0))

xx = np.linspace(-1,4,101)

# Plotting
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(5*phi,5))

plt.plot(xx,f_l(xx))
plt.plot(xx,df_l(xx),color='tab:green')
plt.plot(xx,id_l(xx),':',color='tab:red',linewidth=3)
plt.plot(xx,id_l_c(xx),':',color='tab:orange',linewidth=3)

plt.xlim(min(xx),max(xx))
plt.legend(['f(x)','f\'(x)','F(x)', 'F(x) - C'])
plt.xlabel('x')
plt.ylabel('f(x) / f\'(x) / F(x)')
plt.title('A function and its inverse integral')

plt.savefig('fig12_codechallenge_12_exercise_0d.png')
plt.show()
files.download('fig12_codechallenge_12_exercise_0d.png')
```
