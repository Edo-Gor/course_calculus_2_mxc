```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Compute the area between two functions

# Functions
x = sym.symbols('x')

f = sym.sqrt(x)
g = x

# Intersection of f and g
intersection_points = sym.solve(f-g)
display(Math('\\text{Intersection points: }' + sym.latex(intersection_points))), print()

# Area between functions
area_f = sym.integrate(f, (x,intersection_points[0],intersection_points[1]) )
area_g = sym.integrate(g, (x,intersection_points[0],intersection_points[1]) )
area_between = sym.integrate(f-g, (x,intersection_points[0],intersection_points[1]) )

display(Math('\\text{Area of f(x): }' + sym.latex(area_f))), print()
display(Math('\\text{Area of g(x): }' + sym.latex(area_g))), print()
display(Math('\\text{Area between f(x) and g(x) = }' + sym.latex(area_between))), print()

# Lambdify
f_l = sym.lambdify(x,f)
g_l = sym.lambdify(x,g)

# Evaluate the functions
xx = np.linspace(0,1.5,201)
x_area = np.linspace(float(intersection_points[0]),float(intersection_points[1]),201)

# Plotting
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))
plt.plot(xx,f_l(xx),linewidth=2,label=r'$f(x) = %s$' %sym.latex(f))
plt.plot(xx,g_l(xx),linewidth=2,label=r'$g(x) = %s$' %sym.latex(g))
plt.fill_between(x_area,f_l(x_area),g_l(x_area),color='k',alpha=.15)
plt.xlim(min(xx),max(xx))
plt.ylim(min(f_l(xx)),max(g_l(xx)))

plt.xlabel('$x$')
plt.ylabel('$f(x)$ or $g(x)$')
plt.title('Area between two functions')
plt.legend()

plt.savefig('fig1_codechallenge_70_exercise_0a.png')
plt.show()
files.download('fig1_codechallenge_70_exercise_0a.png')
```

```python
# %% Exercise 0b
#    Compute the area between two functions

# Functions
x = sym.symbols('x')

f = sym.sin(x*10) + sym.log(x+.1) - (x-1)**2 + 3
g = x**2 + sym.cos(x*8)

# Intersection of f and g
bounds = [0.5,1.4]

# Area between functions
area_f = sym.integrate(f, (x,bounds[0],bounds[1]) )
area_g = sym.integrate(g, (x,bounds[0],bounds[1]) )
area_between = sym.integrate(g-f, (x,bounds[0],bounds[1]) )

display(Math('\\text{Area of f(x): }' + sym.latex(area_f))), print()
display(Math('\\text{Area of g(x): }' + sym.latex(area_g))), print()
display(Math('\\text{Area between f(x) and g(x) = }' + sym.latex(area_between))), print()

# Lambdify
f_l = sym.lambdify(x,f)
g_l = sym.lambdify(x,g)

# Evaluate the functions
xx = np.linspace(0,1.5,201)
x_area = np.linspace(float(bounds[0]),float(bounds[1]),201)

# Plotting
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))
plt.plot(xx,f_l(xx),linewidth=2,label=r'$f(x) = %s$' %sym.latex(f))
plt.plot(xx,g_l(xx),linewidth=2,label=r'$g(x) = %s$' %sym.latex(g))
plt.fill_between(x_area,f_l(x_area),g_l(x_area),color='k',alpha=.15)
plt.xlim(min(xx),max(xx))

plt.xlabel('$x$')
plt.ylabel('$f(x)$ or $g(x)$')
plt.title('Area between two functions')
plt.legend()

plt.savefig('fig2_codechallenge_70_exercise_0b.png')
plt.show()
files.download('fig2_codechallenge_70_exercise_0b.png')
```
