```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 1
#    Compute the antiderivative shown in the video using sympy and print

x = sym.symbols('x')
C = sym.symbols('C')

f = x**2
p = sym.pi

display(Math(r'f(x) = %s' % sym.latex(p*f))), print()

F1 = sym.integrate(p*f,x) + C
display(Math(r'F(x) = %s' % sym.latex(F1))), print()

F2 = p * (sym.integrate(f,x) + C)
display(Math(r'F(x) = %s' % sym.latex(F2)))
```

```python
# %% Exercise 2
#    Compute the definite integral shown in the video using sympy and print

a = 0
b = 3*sym.pi

int1 = sym.integrate(f*p,(x,a,b))
int2 = p * sym.integrate(f,(x,a,b))

display(Math('%s = %s' %(sym.latex(sym.Integral(f*p,(x,a,b))),sym.latex(int1))))
display(Math('%s%s = %s' %(sym.latex(p),sym.latex(sym.Integral(f,(x,a,b))),sym.latex(int2))))
```

```python
# %% Exercise 3a
#    Reproduce the plot shown in the video using

# Lambdify the function
f_lamb1 = sym.lambdify(x,f)
f_lamb2 = sym.lambdify(x,f*p)

xx = np.linspace(float(a)-np.pi/2,float(b)+np.pi/2,251)
bounds = np.linspace(float(a),float(b),251)

# Generate x-axis ticks and labels in radians
numTicks     = int((b-a) / (sym.pi/2))
xtick_labels = []
xtick_vals   = []
for i in range(numTicks):
  xtick_vals.append(a+i*(np.pi/2))
  xtick_labels.append('$%s$'%sym.latex(a+i*(sym.pi/2)))


# Plotting
phi = (1 + np.sqrt(5)) / 2
_,ax = plt.subplots(1,figsize=(5*phi,5))

ax.plot(xx,f_lamb1(xx),label=r'$f(x) = %s$'%sym.latex(f),color='tab:red',linewidth=1)
ax.fill_between(bounds,f_lamb1(bounds),alpha=.2,color='tab:red')
ax.plot(xx,np.ones(len(xx))*f_lamb2(xx),label=r'$f(x) = %s$'%sym.latex(p*f),color='tab:blue',linewidth=1)
ax.fill_between(bounds,f_lamb2(bounds),color='none', hatch='X', edgecolor='tab:blue',zorder=-10)
plt.axhline(0,linestyle=':',color='grey',zorder=-4)
plt.title("On the linearity of integration")

ax.set(xlabel='x',ylabel='y',xlim=xx[[0,-1]],xticks=xtick_vals,xticklabels=xtick_labels)
ax.legend()

plt.tight_layout()

plt.savefig('fig8_codechallenge_39_exercise_3a.png')
plt.show()
files.download('fig8_codechallenge_39_exercise_3a.png')
```
