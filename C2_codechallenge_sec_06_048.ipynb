```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Write a function to determine the symmetry of a function

# Functions
x  = sym.symbols('x')
f1 = x**2
f2 = x**3
f3 = x**2 + x**3

def is_even_or_odd(fx):

    # Even if f(x) = f(-x)
    is_even = sym.simplify(fx.subs(x,-x) - fx) == 0

    # Odd if f(-x) = -f(x)
    is_odd = sym.simplify(fx.subs(x,-x) + fx) == 0

    # Return result
    if is_even:
        result = 'even'
    elif is_odd:
        result = 'odd'
    else:
        result = 'neither even nor odd'

    return result

for ff in (f1,f2,f3):
    display(Math('f(x) = %s \\text{ is } \\text{%s}.' %(sym.latex(ff),is_even_or_odd(ff))))
```

```python
# %% Exercise 0b
#    Test symmetry of more functions and compute integrals between the given
#    bounds

# Functions and bounds
# f = -4*x**5 + x**3 - 10*x
f = sym.cos(3*x) + (x-0)**2 - 3

a = -3
b = np.abs(a)

# Integrals
int_a2b = sym.integrate(f,(x,a,b))
int_02b = sym.integrate(f,(x,0,b))

# Print
display(Math('\\text{This function is %s}.' %is_even_or_odd(f)))
display(Math('%s = %g' %(sym.latex(sym.Integral(f,(x,a,b))),int_a2b)))
display(Math('%s = %g' %(sym.latex(sym.Integral(f,(x,0,b))),int_02b)))
display(Math('2%s = %g' %(sym.latex(sym.Integral(f,(x,0,b))),2*int_02b)))
```

```python
# %% Exercise 0c
#    Visualise

# Lambdify the function
xx = np.linspace(a-.25,b+.25,501)
f_lamb = sym.lambdify(x,f)

phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

# Plotting
plt.plot(xx,f_lamb(xx),'k',linewidth=2,label=r'$f(x) = %s$' %sym.latex(f))
plt.axhline(0,color='gray',linestyle=':',zorder=-3)
plt.axvline(0,color='gray',linestyle=':',zorder=-3)
plt.axvline(a,color='gray',linestyle=':',zorder=-3)
plt.axvline(b,color='gray',linestyle=':',zorder=-3)

plt.plot(a,f_lamb(a),'o',markersize=8,label=r'$f(%g)=%g$' %(a,f_lamb(a)))
plt.plot(b,f_lamb(b),'o',markersize=8,label=r'$f(%g)=%g$' %(a,f_lamb(b)))

# Find x values for this segment and draw patch
xx_segment = xx[(xx>a) & (xx<0)]
plt.fill_between(xx_segment,f_lamb(xx_segment),color='b',alpha=.2)
xx_segment = xx[(xx>=0) & (xx<b)]
plt.fill_between(xx_segment,f_lamb(xx_segment),color='m',alpha=.2)

plt.title("Definite integral of a symmetric function")
plt.gca().set(xlim=xx[[0,-1]],ylim=[np.min(f_lamb(xx))-1,np.max(f_lamb(xx))+1],xlabel='x',ylabel='y = f(x)')
plt.legend()

plt.savefig('fig20_codechallenge_48_exercise_0c.png')
plt.show()
files.download('fig20_codechallenge_48_exercise_0c.png')
```
