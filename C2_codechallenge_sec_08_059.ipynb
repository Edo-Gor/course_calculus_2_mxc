```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Reproduce the integral shown in the video

# Function
x  = sym.symbols('x')
fx = sym.Piecewise((x/2,x<0),(x**2+1,x>=0))
F  = sym.integrate(fx)
df = sym.diff(fx)

a,b = -2,1
sym.integrate(fx,(x,a,b))
display(Math('%s = %s' %(sym.latex(sym.Integral(fx)),sym.latex(sym.integrate(fx)))))

# Plotting
p = sym.plot(fx,(x,-3,2),F,df,title='A function, its integral, and its derivative',ylabel='f(x) / F(x)',xlabel='x',legend=True,label=['Function', 'Integral', 'Derivative'],show=False)
p._backend.process_series()
lines = p._backend.ax.get_lines()
lines[0].set_linewidth(1.5)
lines[1].set_linewidth(1.5)
lines[2].set_linewidth(1.5)
lines[1].set_linestyle('--')
lines[2].set_linestyle(':')

p._backend.fig.savefig('fig6_codechallenge_59_exercise_0a.png')
files.download('fig6_codechallenge_59_exercise_0a.png')
```

```python
# %% Exercise 0b
#    Reproduce the integral shown in the video

# Function
x  = sym.symbols('x')
fx = 1 / sym.sqrt(x-1)
F  = sym.integrate(fx)
df = sym.diff(fx)

a,b = 1,2
sym.integrate(fx,(x,a,b))
display(Math('%s = %s' %(sym.latex(sym.Integral(fx)),sym.latex(sym.integrate(fx)))))

# Plotting
p = sym.plot(fx,(x,1,3),F,df,ylim=[-10,10],title='A function, its integral, and its derivative',ylabel='f(x) / F(x)',xlabel='x',legend=True,label=['Function', 'Integral', 'Derivative'],show=False)
p._backend.process_series()
lines = p._backend.ax.get_lines()
lines[0].set_linewidth(1.5)
lines[1].set_linewidth(1.5)
lines[2].set_linewidth(1.5)
lines[1].set_linestyle('--')
lines[2].set_linestyle(':')

p._backend.fig.savefig('fig7_codechallenge_59_exercise_0b.png')
files.download('fig7_codechallenge_59_exercise_0b.png')
```

```python
# %% Exercise 0c
#    Reproduce the integral shown in the video

# Function
x  = sym.symbols('x')
fx = (x**2-2*x) / (x**2-4)
F  = sym.integrate(fx)
df = sym.diff(fx)

a,b = 0,4
sym.integrate(fx,(x,a,b))
display(Math('%s = %s' %(sym.latex(sym.Integral(fx)),sym.latex(sym.integrate(fx)))))

# Plotting
p = sym.plot(fx,(x,-2,6),F,df,ylim=[-2,2],title='A function, its integral, and its derivative',ylabel='f(x) / F(x)',xlabel='x',legend=True,label=['Function', 'Integral', 'Derivative'],show=False)
p._backend.process_series()
lines = p._backend.ax.get_lines()
lines[0].set_linewidth(1.5)
lines[1].set_linewidth(1.5)
lines[2].set_linewidth(1.5)
lines[1].set_linestyle('--')
lines[2].set_linestyle(':')

p._backend.fig.savefig('fig8_codechallenge_59_exercise_0c.png')
files.download('fig8_codechallenge_59_exercise_0c.png')
```
