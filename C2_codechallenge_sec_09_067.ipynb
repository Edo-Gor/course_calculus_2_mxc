```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 1
#    Reproduce the plots shown in the video

# Preallocate
x = np.linspace(-2,4,50001)

As = np.linspace(0,1,31)
Bs = np.linspace(0,2,15)

Areas = np.zeros((len(As),len(Bs)))

# Plotting
phi = (1 + np.sqrt(5)) / 2
fig,axs = plt.subplots(1,2,figsize=(2*phi*5,5))

cmap = plt.cm.plasma(np.linspace(.1,.9,len(Bs)))

for ai,a in enumerate(As):
  for bi,b in enumerate(Bs):

    # Create the function (unique for this a/b parameter pair)
    fx = lambda xx : np.abs(xx)*np.cos((a+xx)**2)+b

    # Plotting
    axs[0].plot(x,fx(x),linewidth=0.5,color=cmap[bi],alpha=0.6)

    # Get and store area
    Areas[ai,bi] = spi.quad(fx,x[0],x[-1])[0]


# Draw  areas
im = axs[1].imshow(Areas,extent=[Bs[0],Bs[-1],As[-1],As[0]],cmap='plasma')
axs[1].set(xlabel='"b" parameter',ylabel='"a" parameter',title=r'$\int_{-2}^{4}f(x,a,b)\,dx$')
fig.colorbar(im,ax=axs[1],shrink=.65)

axs[0].axis('off')
plt.tight_layout()

plt.savefig('fig6_codechallenge_67_exercise_1.png')
plt.show()
files.download('fig6_codechallenge_67_exercise_1.png')
```

```python
# %% Exercise 2
#    Reproduce the plots shown in the video

# Function
t  = sym.symbols('t')
ft = (t**2+4) / (3*t**3-4*t**2-4*t)

# Take 1st and 2nd integral
F  = sym.integrate(ft)
F2 = sym.integrate(F)

display(Math('%s = %s+C' %(sym.latex(sym.Integral(ft)),sym.latex(F)))),print()
display(Math('%s = %s+C' %(sym.latex(sym.Integral(F)),sym.latex(F2)))),print()


# Lambdify
tt    = np.linspace(-np.pi,2*np.pi,501)
ft_l  = sym.lambdify(t,ft)
if_l  = sym.lambdify(t,F)
if_l2 = sym.lambdify(t,F2)

# Plotting
phi = (1 + np.sqrt(5)) / 2
_,axs = plt.subplots(1,2,figsize=(1.5*phi*5,5))

axs[0].plot(tt,ft_l(tt),'tab:blue',label='f(t)')
axs[0].plot(tt,np.real(if_l(tt.astype(complex))),'tab:green',label=r'$re\left(\int\right)$')
axs[0].plot(tt,np.imag(if_l(tt.astype(complex))),'--',color='tab:green',label=r'$im\left(\int\right)$')
axs[0].plot(tt,np.real(if_l2(tt.astype(complex))),'tab:red',label=r'$re\left(\int\int\right)$')
axs[0].plot(tt,np.imag(if_l2(tt.astype(complex))),'--',color='tab:red',label=r'$im\left(\int\int\right)$')
axs[0].set(xlim=tt[[0,-1]],ylim=[-10,10],xlabel='t',ylabel='y=f(t) / y=F(t)',title='Function and integrals in x-space')
axs[0].legend(ncol=2,fontsize=10)

axs[1].plot(np.real(if_l(tt.astype(complex))),np.imag(if_l(tt.astype(complex))),'.',color='tab:green',linewidth=2)
axs[1].plot(np.real(if_l2(tt.astype(complex))),np.imag(if_l2(tt.astype(complex))),'.',color='tab:red',linewidth=2)
axs[1].set(xlabel='Real part',ylabel='Imaginary part',title='Definite integrals in complex plane')

plt.tight_layout()

plt.savefig('fig8_codechallenge_67_exercise_2.png')
plt.show()
files.download('fig8_codechallenge_67_exercise_2.png')
```
