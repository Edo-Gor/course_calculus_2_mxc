```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import scipy.integrate   as spi
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Compute the surface of solids of revolution

# Function
x  = sym.symbols('x')
fx = x**2 + sym.cos(x**2)
df = sym.diff(fx)

# Bounds
a = 0
b = 2

# Get solid surface integral
integrand = fx * sym.sqrt( 1 + df**2 )

# Print out the problem
display(Math('f(x) \\;=\\; %s' % (sym.latex(fx))))
display(Math('A \\;=\\; 2\\pi %s' % (sym.latex(sym.Integral(integrand,(x,a,b))))))

# Try to get the analytical solution (takes a few mins)
surf_area = 2*sym.pi * sym.integrate(integrand,(x,a,b))
print(f'The surface area of the solid is: {surf_area.evalf():.5f}')
```

```python
# %% Exercise 0a
#    Continue ...

# Redefine the function (alternative to lambdifying)
def f(x): return x**2 + np.cos(x**2)

# Theta values for revolution and meshgrid
theta   = np.linspace(0,2*np.pi,101)
xx      = np.linspace(a,b,103)
X,Theta = np.meshgrid(xx,theta)

# Y and Z coordinates for f(x)
Y_f = f(X) * np.cos(Theta)
Z_f = f(X) * np.sin(Theta)


# Plot
phi = (1 + np.sqrt(5)) / 2
fig = plt.figure(figsize=(phi*8,8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y_f, Z_f, color='blue', alpha=.8, edgecolor='none')

ax.set(xlabel='X',ylabel='Y',zlabel='Z')
ax.set_title(f'Solid of revolution (surface area = {surf_area:.2f} a.u.$^2$)')

plt.savefig('fig37_codechallenge_80_exercise_0a.png')
plt.show()
files.download('fig37_codechallenge_80_exercise_0a.png')
```
