```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# %% Exercise 0a
#    Function setup

def f(u):
    return u**2 - 0.5

# Grid spacing
dx = 0.02
xx = np.arange(-1,1+dx,dx)
yy = f(xx)

# Table of x,y pairs
print("   x   |   y   ")
print("-------|-------")
for xi,yi in zip(xx,yy):
    print(f"{xi:6.3f} | {yi:6.3f}")
```

```python
# %% Exercise 0b
#    Plotting

# Plot function
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(5*phi,5))

plt.plot(xx,yy,'ks-',linewidth=2,markersize=10,markerfacecolor="tab:blue")
cmap = plt.cm.plasma(np.linspace(.1,.9,len(xx)))

# Initialise area
area = 0

# Plot recatngles
for xi,i in enumerate(xx):

    plt.fill_between([i-dx/2,i+dx/2],[f(i),f(i)],edgecolor="k",facecolor=cmap[xi],alpha=0.8)
    area += f(i)*dx

plt.xlim(min(xx)-dx/3,max(xx)+dx/3)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(f"Area = {area:.3f}")

plt.savefig('fig6_codechallenge_09_exercise_0b.png')
plt.show()
files.download('fig6_codechallenge_09_exercise_0b.png')
```
