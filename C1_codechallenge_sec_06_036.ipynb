```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math
import mpmath

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 1
#    Reproduce the plot shown in the video

# Variables
x = sym.symbols('x',real=True)
C = sym.symbols('C',real=True)

# Variables and problem constraints (x_0 and f(x_0))
df = 2*x + 3
constraints = [1,2]

# Integrate to find f(x)
f = sym.integrate(df,x) + C

# Solve for C
constant = sym.solve(f.subs(x,constraints[0]) - constraints[1],C)[0]

# Print
display(Math('\\text{PROBLEM:}'))
display(Math("\\text{Given } f'(x) = %s \\text{, and } f(%g)=%g" %(sym.latex(df),constraints[0],constraints[1])))
print()
display(Math('\\text{SOLUTION:}'))
display(Math("f(x) = " + sym.latex(f.subs(C,constant))))
```

```python
# %% Exercise 1
#    Continue ...

# Plotting (remove negative domain if imaginary)
xx = np.linspace(-2,2,41)

# Parallel curves
wrongConstants = np.linspace(float(constant)-2,float(constant)+2,12)

for ci in wrongConstants:

  y = [f.subs(C,ci).subs(x,i) for i in xx]
  plt.plot(xx,y,color=[.6,.6,.6],linewidth=.4)

y = [f.subs(C,constant).subs(x,i) for i in xx]
plt.plot(xx,y,color=[.5,0,.6],linewidth=5)

plt.title('Initial value problem for integration contant')
plt.xlim(xx[[0,-1]])

plt.savefig('fig5_codechallenge_35_exercise_1.png')
plt.show()
files.download('fig5_codechallenge_35_exercise_1.png')
```
