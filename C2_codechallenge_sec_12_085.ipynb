```python
import numpy                as np
import sympy                as sym
import matplotlib.pyplot    as plt
import matplotlib.colors    as mcolors
import scipy.integrate      as spi
import math
import mpmath
import plotly.graph_objects as go
import sympy.stats

from scipy.signal                     import find_peaks
from scipy                            import stats
from IPython.display                  import display,Math
from google.colab                     import files
from IPython.display                  import Audio
from scipy.io                         import wavfile
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')

import matplotlib.animation as animation
from matplotlib import rc
rc('animation', html='jshtml')
```

```python
# %% Exercise 0a
#    Pdfs using sympy

# Parameters
x  = sym.symbols('x')
m  = 1.5
a  = 2
xx = np.linspace(.01,5,500)

# Sympy expression (Pareto distribution)
P         = sym.stats.Pareto('P',m,a)
Pcdf_expr = sym.stats.cdf(P)(x)
cdf_expr  = sym.diff(Pcdf_expr,x)

display(Math('cdf \\;=\\; %s' %sym.latex(Pcdf_expr)))
print()

# Lambdify and evaluate numerically
Pcdf_l = sym.lambdify(x,Pcdf_expr)
cdf_l  = sym.lambdify(x,cdf_expr)

cdf      = Pcdf_l(xx)
cdf_diff = cdf_l(xx)

# Print some values
display(Math('c(%g) \\;=\\; %g' %(xx[-1],cdf[-1])))
display(Math('c(%s) \\;=\\; %g' %(sym.latex(sym.oo),Pcdf_l(sym.oo))))
print()

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

plt.plot(xx,cdf,label='CDF function')
plt.plot(xx,cdf_diff,'--',label='CDF derivative (pdf)')
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

plt.gca().set(xlim=xx[[0,-1]],xlabel='x',ylabel='Probability',title=f'Pareto cdf ($m$ = {m}, $\\alpha$ = {a})')
plt.legend()
plt.tight_layout()

plt.savefig('fig8_codechallenge_85_exercise_0a.png')
plt.show()
files.download('fig8_codechallenge_85_exercise_0a.png')
```

```python
# %% Exercise 0b
#    Pdfs using scipy

# Parameters
mu    = 1.0
sigma = 0.3

# Compute pdf values
xx  = np.linspace(-5,5,401)
cdf = stats.logistic.cdf(xx,loc=mu,scale=sigma)
cdf_diff = stats.logistic.pdf(xx,loc=mu,scale=sigma)

# Plot
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(phi*5,5))

plt.plot(xx,cdf,label='CDF function')
plt.plot(xx,cdf_diff,'--',label='CDF derivative (pdf)')
plt.axvline(mu,color='gray',linestyle=':',linewidth=0.8)
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

plt.gca().set(xlim=xx[[0,-1]],xlabel='x',ylabel='Cumulative prob.',title=r'Logistic cdf ($\mu = %g, \sigma = %g$)' %(mu,sigma))
plt.legend()
plt.tight_layout()

plt.savefig('fig9_codechallenge_85_exercise_0b.png')
plt.show()
files.download('fig9_codechallenge_85_exercise_0b.png')
```
