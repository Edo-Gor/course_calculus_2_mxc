```python
import numpy             as np
import sympy             as sym
import matplotlib.pyplot as plt
import math

from scipy.signal                     import find_peaks
from IPython.display                  import display,Math
from google.colab                     import files
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('svg')
```

```python
# %% Exercise 0a
#    Show that the definite integral is the difference between the indefinite
#    integral evaluated at the bounds

x,C = sym.symbols('x,C')
f   = (x-1)**2
a   = -0.5
b   = sym.pi

def_int = sym.integrate(f,(x,a,b))
ind_int = sym.integrate(f,x)

ind_a = ind_int.subs(x,a)
ind_b = ind_int.subs(x,b)

display(Math("\\text{Function: } %s" %sym.latex(f))), print()
display(Math('%s = %s' %(sym.latex(sym.Integral(f,x)),sym.latex(ind_int)))), print()
display(Math('F(%s) = %s' %(sym.latex(a),sym.latex(ind_a)))), print()
display(Math('F(%s) = %s' %(sym.latex(b),sym.latex(ind_b.evalf())))), print()
display(Math('F(%s)-F(%s) = %s' %(sym.latex(b),sym.latex(a),sym.latex((ind_b-ind_a).evalf())))), print()
display(Math('\\int_{%s}^{%s} %s dx = %s' %(sym.latex(a),sym.latex(b),sym.latex(f),sym.latex(def_int.evalf()))))
```

```python
# %% Exercise 0b
#    Visualise

# Lambdify
f_l   = sym.lambdify(x,f,'numpy')
int_l = sym.lambdify(x,ind_int,'numpy')
Fa    = int_l(float(a))
Fb    = int_l(float(b))

xx = np.linspace(-1,4,101)

# Plotting
phi = (1 + np.sqrt(5)) / 2
plt.figure(figsize=(5*phi,5))

plt.plot(xx,f_l(xx),label=r'$f(x) = %s$'%sym.latex(f))
plt.plot(xx,int_l(xx),label=r'$F(x) = %s$'%sym.latex(ind_int))

plt.axvline(x=a,color='r',linestyle=':',linewidth=0.8,label=r'$x_a$ = %.2f'%a)
plt.axvline(x=b,color='r',linestyle=':',linewidth=0.8,label=r'$x_b$ = %.2f'%b)
plt.text(a+.03,f_l(a)*1.3,r'$F(%.2f) = %.2f$' %(a,Fa),rotation=90)
plt.text(b-0.01,f_l(b)*1.1,r'$F(%.2f) = %.2f$' %(b,Fb),rotation=90,ha='right')

plt.axvline(x=0,color='grey',linestyle=':',linewidth=0.8)
plt.axhline(y=0,color='grey',linestyle=':',linewidth=0.8)

x4area = np.linspace(float(a),float(b),99)
plt.fill_between(x4area,f_l(x4area),label=f'Area = {Fb-Fa:.2f}',color='tab:blue',edgecolor=None,alpha=.2)

plt.xlim(min(xx),max(xx))
plt.legend()
plt.xlabel('x')
plt.ylabel('f(x) / F(x)')
plt.title('A function and its integral')

plt.savefig('fig13_codechallenge_13_exercise_0b.png')
plt.show()
files.download('fig13_codechallenge_13_exercise_0b.png')
```
