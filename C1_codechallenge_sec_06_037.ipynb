```python
# %% Exercise 1
#    Generate antiderivatives of polynomials problems using 2 to 4 terms,
#    integer coefficients between -10 and 10, randomly scale odd powers by 2,
#    randomly scale even powers by -1

# Variables and terms
x = sym.symbols('x')
C = sym.symbols('C')

terms = np.random.randint(2,5)
betas = np.random.randint(-10,11,terms)

# Build function to integrate
f = 0

for i,c in enumerate(betas):

    if i%2 == 1:
        scalar = np.random.choice([1,2],1)[0]

    else:
        scalar = np.random.choice([-1,1],1)[0]

    f += c*x**sym.Rational(i,scalar)

display(Math("\\text{Determine the following antiderivative:}"))
print()
display(Math(sym.latex(sym.Integral(f))))
```

```python
# %% Exercise 1
#    Continue ...

answer = sym.integrate(f,x) + C

display(Math("\\text{The antiderivative is:}"))
print()
display(Math(sym.latex(answer)))
```

```python
# %% Exercise 2
#    Same as in exercise 1 but with definite integrals with random integer
#    bounds between -3 and 3, if the function contains radicals, the lower bound
#    should be adjusted so that it is positive

# Variables and terms
x = sym.symbols('x')
C = sym.symbols('C')

terms = np.random.randint(2,5)
betas = np.random.randint(-10,11,terms)

# Bounds (must be sorted and unequal)
a,b = np.sort(np.random.randint(-3,4,2))
if a==b: b+=1

# Build function to integrate
f = 0

for i,c in enumerate(betas):

    if i%2==1:
        scalar = np.random.choice([1,2],1)[0]

    else:
        scalar = np.random.choice([-1,1],1)[0]

    f += c * x**sym.Rational(i,scalar)


# Eradicate negative radicals
radical = 0

for term in f.as_ordered_terms():
    try:
        radical += not term.args[-1].args[-1].is_integer
    except:
        pass

if radical>0:
    a,b = np.sort(np.abs([a,b]))

display(Math("\\text{Compute the following definite integral:}"))
print()
display(Math(sym.latex(sym.Integral(f))))
```

```python
# %% Exercise 2
#    Continue ...

answer = sym.integrate(f,(x,a,b))

display(Math("\\text{The definite integral equals:}"))
print()
display(Math(sym.latex(answer)))
```
